```python
import streamlit as st
import cv2
import numpy as np
from ultralytics import YOLO
import mediapipe as mp
from collections import Counter
import os

# 🚀 Title
st.title("📹 Real-time Exam Proctoring System")
st.markdown("This system detects **phones**, **people**, and uses face detection for proctoring.")

# 📦 Load YOLOv8
st.write("📦 Loading YOLOv8 model...")
yolo_model = YOLO('yolov8m.pt')

# 📦 Load MobileNet-SSD (absolute paths)
st.write("📦 Loading MobileNet-SSD model...")
proto_path = r"C:\Users\sagni\ultralytics\deploy.prototxt.txt"
model_path = r"C:\Users\sagni\ultralytics\mobilenet_iter_73000.caffemodel"

# Check if files exist
if not os.path.exists(proto_path):
    st.error(f"❌ deploy.prototxt.txt not found at {proto_path}")
if not os.path.exists(model_path):
    st.error(f"❌ mobilenet_iter_73000.caffemodel not found at {model_path}")

# Load model
ssd_net = cv2.dnn.readNetFromCaffe(proto_path, model_path)

# 📦 Initialize MediaPipe Face Detection
st.write("📦 Initializing MediaPipe Face Detection...")
mp_face = mp.solutions.face_detection
face_detector = mp_face.FaceDetection(model_selection=1, min_detection_confidence=0.5)

# 🎥 Start webcam
st.write("🎥 Starting webcam...")
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    st.error("❌ Cannot open webcam. Please check your camera and try again.")

# Stats Counter
stats = Counter()

# 🧠 Define helper function for people detection
def detect_people(frame):
    h, w = frame.shape[:2]
    blob = cv2.dnn.blobFromImage(frame, 0.007843, (300, 300), 127.5)
    ssd_net.setInput(blob)
    detections = ssd_net.forward()
    people_boxes = []

    for i in range(detections.shape[2]):
        confidence = detections[0, 0, i, 2]
        class_id = int(detections[0, 0, i, 1])
        if confidence > 0.5 and class_id == 15:  # Class 15 = person
            box = detections[0, 0, i, 3:7] * np.array([w, h, w, h])
            (x1, y1, x2, y2) = box.astype("int")
            people_boxes.append((x1, y1, x2, y2))
    return people_boxes

# 🔁 Real-time processing loop
stframe = st.empty()  # Placeholder for video frames
while True:
    ret, frame = cap.read()
    if not ret:
        st.error("❌ Failed to grab frame")
        break

    # 📱 Detect phones with YOLOv8
    results = yolo_model(frame, imgsz=640, verbose=False)
    for r in results:
        for box in r.boxes.xyxy:
            x1, y1, x2, y2 = map(int, box)
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
            cv2.putText(frame, "Phone", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 0), 2)
            stats['phones'] += 1

    # 👥 Detect people with MobileNet-SSD
    people_boxes = detect_people(frame)
    stats['people'] += len(people_boxes)
    for (x1, y1, x2, y2) in people_boxes:
        cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
        cv2.putText(frame, "Person", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)

    # 🙂 Detect faces with MediaPipe
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    results = face_detector.process(rgb_frame)
    if results.detections:
        for detection in results.detections:
            bbox = detection.location_data.relative_bounding_box
            h, w, _ = frame.shape
            x1, y1, w_box, h_box = int(bbox.xmin * w), int(bbox.ymin * h), int(bbox.width * w), int(bbox.height * h)
            cv2.rectangle(frame, (x1, y1), (x1 + w_box, y1 + h_box), (0, 255, 255), 2)
            cv2.putText(frame, "Face", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0, 255, 255), 2)

    # 📝 Show stats
    cv2.putText(frame, f'Phones Detected: {stats["phones"]}', (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
    cv2.putText(frame, f'People Detected: {stats["people"]}', (10, 60), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

    # 🖼 Display in Streamlit
    stframe.image(frame, channels="BGR")

# Cleanup
cap.release()
st.success("✅ Stream ended")
```

```python
import cv2
import numpy as np
from ultralytics import YOLO
import mediapipe as mp
from matplotlib import pyplot as plt

# Load YOLOv8 for phone detection
print("📦 Loading YOLOv8 model...")
yolo_model = YOLO('yolov8m.pt')

# Load Haar Cascade for people detection (face fallback)
print("📦 Loading Haar Cascade for face detection...")
haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Initialize MediaPipe Face Detection for exclusion
print("📦 Initializing MediaPipe Face Detection...")
mp_face = mp.solutions.face_detection
face_detector = mp_face.FaceDetection(model_selection=1, min_detection_confidence=0.5)

# Statistics
stats = {'phones': 0, 'people': 0}

# Start video capture
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("❌ Cannot open camera. Check permissions or try changing VideoCapture index (0 → 1).")
    cap.release()
else:
    print("🚀 Starting real-time detection (Stop kernel or press 'Ctrl+C' to quit)...")
    plt.ion()
    fig, ax = plt.subplots(figsize=(8, 6))

    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                print("❌ Failed to grab frame. Is your camera in use by another app?")
                break

            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Detect phones with YOLOv8
            results = yolo_model(frame_rgb, verbose=False)
            for result in results:
                for box, cls in zip(result.boxes.xyxy, result.boxes.cls):
                    x1, y1, x2, y2 = map(int, box)
                    class_name = yolo_model.names[int(cls)]
                    if class_name == "cell phone":
                        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
                        cv2.putText(frame, "Phone", (x1, y1 - 10),
                                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
                        stats['phones'] += 1

            # Detect people with Haar Cascade (faces as proxy)
            gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
            faces = haar_cascade.detectMultiScale(gray, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
            for (x, y, w, h) in faces:
                cv2.rectangle(frame, (x, y), (x + w, y + h), (255, 0, 0), 2)
                cv2.putText(frame, "Person", (x, y - 10),
                            cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)
                stats['people'] += 1

            # Display statistics
            cv2.putText(frame, f'Phones: {stats["phones"]}', (10, 30),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (0, 255, 0), 2)
            cv2.putText(frame, f'People: {stats["people"]}', (10, 60),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 0, 0), 2)

            # Show frame using Matplotlib
            ax.clear()
            ax.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
            ax.axis('off')
            plt.pause(0.001)

    except KeyboardInterrupt:
        print("👋 Exiting detection...")

    cap.release()
    plt.ioff()
    plt.show()
```
