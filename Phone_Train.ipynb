```python
import cv2
import os

# Folder to save images
folder = r"C:\Users\sagni\Downloads\Face Concentration\Phone_Dataset\front_phones"
os.makedirs(folder, exist_ok=True)

# Start webcam
cap = cv2.VideoCapture(0)
print("üì± Capturing FRONT phones. Press SPACE to save, 'q' to quit.")

counter = 1
while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ùå Camera error")
        break

    frame = cv2.flip(frame, 1)
    cv2.imshow("Front Phone Capture", frame)

    key = cv2.waitKey(1)
    if key == ord(' '):  # SPACE to save
        filename = os.path.join(folder, f"front_phone_{counter:03}.jpg")
        cv2.imwrite(filename, frame)
        print(f"‚úÖ Saved {filename}")
        counter += 1
    elif key == ord('q'):  # q to quit
        break

cap.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import os

# Paths
dataset_root = r"C:\Users\sagni\Downloads\Face Concentration\Phone_Dataset"
output_labels = os.path.join(dataset_root, "labels_auto")
os.makedirs(output_labels, exist_ok=True)

# Load YOLOv8 pre-trained model (or your custom model)
model = YOLO('yolov8x.pt')  # Replace with your custom model path if needed

# Loop through all images in dataset
for root, _, files in os.walk(dataset_root):
    for file in files:
        if file.lower().endswith(('.jpg', '.png')):
            img_path = os.path.join(root, file)
            img = cv2.imread(img_path)

            # Predict objects
            results = model.predict(img, imgsz=640, conf=0.2, verbose=False)[0]

            # YOLO format label file path
            rel_path = os.path.relpath(root, dataset_root)
            label_folder = os.path.join(output_labels, rel_path)
            os.makedirs(label_folder, exist_ok=True)
            label_file = os.path.join(label_folder, file.rsplit('.', 1)[0] + '.txt')

            with open(label_file, 'w') as f:
                for box in results.boxes:
                    cls_id = int(box.cls[0])
                    conf = float(box.conf[0])
                    x1, y1, x2, y2 = box.xyxy[0]

                    # Normalize coordinates (YOLO format)
                    img_h, img_w = img.shape[:2]
                    x_center = ((x1 + x2) / 2) / img_w
                    y_center = ((y1 + y2) / 2) / img_h
                    width = (x2 - x1) / img_w
                    height = (y2 - y1) / img_h

                    # Filter for phone-related classes (class names vary per model)
                    class_name = model.names[cls_id].lower()
                    if 'phone' in class_name or 'cell' in class_name or 'mobile' in class_name:
                        print(f"üì± {file} -> {class_name} @ {round(conf, 2)}")
                        f.write(f"0 {x_center} {y_center} {width} {height}\n")
                    elif 'eye' in class_name or 'reflection' in class_name:
                        print(f"üëÅÔ∏è {file} -> {class_name} @ {round(conf, 2)}")
                        f.write(f"1 {x_center} {y_center} {width} {height}\n")
```

```python
from ultralytics import YOLO
import cv2

# üöÄ Load trained model
model = YOLO(r"C:/Users/sagni/Downloads/Face Concentration/models/yolosagnik6/weights/best.pt")

# üñ•Ô∏è Start webcam
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("‚ùå Could not open webcam")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("‚ö†Ô∏è Frame capture failed")
        break

    frame = cv2.flip(frame, 1)  # Mirror view
    results = model.predict(source=frame, imgsz=640, conf=0.3, device="cpu")[0]

    for box in results.boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        cls_id = int(box.cls[0])
        conf = box.conf[0]
        label = f"{model.names[cls_id]} {conf:.2f}"
        color = (0, 255, 0) if cls_id == 0 else (0, 0, 255)  # Green for phone, Red for reflection

        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    cv2.imshow("üì± Phone & Reflection Detector", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
